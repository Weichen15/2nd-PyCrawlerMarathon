```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

'''
Your Code
'''
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

import json
data = json.loads(response)

print("é€™å€‹APIä¸€æ¬¡å‚³å›%sç­†è³‡æ–™" % len(data))
col = []
print("æ¯ç­†è³‡æ–™åŒ…å«çš„æ¬„ä½ï¼š")
for t in data[0]:
    col.append(t)
print(col)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

'''
Your Code
'''
title = []
print("æ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ç‚º:")
for i in range(len(data)):
    title.append(data[i]['title'])
print(title, '\n')

time = []
print("æ¯ä¸€ç­†è³‡æ–™çš„ã€Œè²¼æ–‡æ™‚é–“ã€ç‚º:")
for i in range(len(data)):
    time.append(data[i]['createdAt'])
print(time ,'\n')

com = []
print("æ¯ä¸€ç­†è³‡æ–™çš„ã€Œç•™è¨€æ•¸ã€ç‚º:")
for i in range(len(data)):
    com.append(data[i]['commentCount'])
print(com ,'\n')

like = []
print("æ¯ä¸€ç­†è³‡æ–™çš„ã€ŒæŒ‰è®šæ•¸ã€ç‚º:")
for i in range(len(data)):
    like.append(data[i]['likeCount'])
print(like ,'\n')
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

## ä»¥dcardç¸½ç‰ˆç‚ºä¾‹ï¼Œè³‡æ–™é‡æ¯”è¼ƒå¤§ï¼Œå·®è·è¼ƒæ˜é¡¯
'''
Your Code
'''
def avg(x):
    like = 0
    com = 0
    avg_like = 0
    avg_com = 0
    for i in range(len(x)):
        like += x[i]['likeCount']
        com += x[i]['commentCount']
    avg_like = like / len(x)
    avg_com = com / len(x)
    return avg_like, avg_com

## ç†±é–€

r1 = requests.get('https://www.dcard.tw/_api/forums/dcard/posts?popular=true')
response = r1.text

data1 = json.loads(response)

## éç†±é–€

r2 = requests.get('https://www.dcard.tw/_api/forums/dcard/posts?popular=false')
response = r2.text

data2 = json.loads(response)

print("ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€ç‚ºï¼š %.1f,\n" % avg(data1)[1],"ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€ç‚ºï¼š %.1f \n"  % avg(data1)[0])
print("éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€ç‚ºï¼š %.1f,\n" % avg(data2)[1],"ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€ç‚ºï¼š %.1f \n"  % avg(data2)[0])
```
