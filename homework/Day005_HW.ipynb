```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

'''
Your Code
'''
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

import json
data = json.loads(response)

print("這個API一次傳回%s筆資料" % len(data))
col = []
print("每筆資料包含的欄位：")
for t in data[0]:
    col.append(t)
print(col)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

'''
Your Code
'''
title = []
print("每一筆資料的「標題」為:")
for i in range(len(data)):
    title.append(data[i]['title'])
print(title, '\n')

time = []
print("每一筆資料的「貼文時間」為:")
for i in range(len(data)):
    time.append(data[i]['createdAt'])
print(time ,'\n')

com = []
print("每一筆資料的「留言數」為:")
for i in range(len(data)):
    com.append(data[i]['commentCount'])
print(com ,'\n')

like = []
print("每一筆資料的「按讚數」為:")
for i in range(len(data)):
    like.append(data[i]['likeCount'])
print(like ,'\n')
```

```python
import pandas as pd
df1 = pd.DataFrame({'標題': title,
                    '貼文時間': time,
                    '留言數': com,
                    '按讚數': like})
df1
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

## 以dcard總版為例，資料量比較大，差距較明顯
'''
Your Code
'''
def avg(x):
    like = 0
    com = 0
    avg_like = 0
    avg_com = 0
    for i in range(len(x)):
        like += x[i]['likeCount']
        com += x[i]['commentCount']
    avg_like = like / len(x)
    avg_com = com / len(x)
    return avg_like, avg_com

## 熱門

r1 = requests.get('https://www.dcard.tw/_api/forums/dcard/posts?popular=true')
response = r1.text

data1 = json.loads(response)

## 非熱門

r2 = requests.get('https://www.dcard.tw/_api/forums/dcard/posts?popular=false')
response = r2.text

data2 = json.loads(response)

print("熱門文章的「平均留言人數」為： %.1f,\n" % avg(data1)[1],"「平均按讚人數」為： %.1f \n"  % avg(data1)[0])
print("非熱門文章的「平均留言人數」為： %.1f,\n" % avg(data2)[1],"「平均按讚人數」為： %.1f \n"  % avg(data2)[0])
```
