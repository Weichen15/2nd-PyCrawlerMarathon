## 作業目標

根據範例 ，完成以下問題：

* ① 取出今天所有的發文
* ② 如果想要依照類別分類，怎麼儲存會比較好？
* ③ 哪一個類別的文章最多




```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")


```

```python
soup
```

### ① 取出今天所有的發文

```python
'''
Your Code
'''
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''
category = []
for d in soup.find(class_="part_list_2").find_all('h3'):
#     print(d.find(class_="date").text, d.find_all('a')[-1].text)
    
        category.append(d.find_all('em')[-1].text)
```

```python
category
```

```python
cat = np.array(category)
cat
df = pd.DataFrame({"Category":cat})
df
```

### ③ 哪一個類別的文章最多

```python
'''
Your Code
'''
a = pd.DataFrame(df["Category"].value_counts(sort = True, ascending = False))
a.head()
```

```python
print("文章最多的類別為：", a.index[0])
```
